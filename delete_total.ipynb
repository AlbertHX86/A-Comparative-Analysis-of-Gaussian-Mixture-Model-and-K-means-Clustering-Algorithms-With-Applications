```python
#清洗数据
import pandas as pd
df = pd.read_csv('Electric_Vehicle_Charging_Station_Energy_Consumption.csv')
df
```

```python
df['time_only_start'] = pd.to_datetime(df['Start_Date___Time']).dt.time
df['time_only_end'] = pd.to_datetime(df['End_Date___Time']).dt.time
df
```

```python
df.drop('Start_Date___Time', axis=1, inplace=True)
df.drop('End_Date___Time', axis=1, inplace=True)

df.drop('Unnamed: 1', axis=1, inplace=True)
df
```

```python
def time_to_seconds(time_str):
    hours, minutes, seconds = map(int, time_str.split(':'))
    total_seconds = hours * 3600 + minutes * 60 + seconds
    return total_seconds


# Sample DataFrame
df['Total_Duration_seconds'] = df['Total_Duration__hh_mm_ss_'].apply(
    time_to_seconds)
df['Charging_Time_seconds'] = df['Charging_Time__hh_mm_ss_'].apply(
    time_to_seconds)
```

```python
df.drop('Total_Duration__hh_mm_ss_', axis=1, inplace=True)
df.drop('Charging_Time__hh_mm_ss_', axis=1, inplace=True)
df
```

```python
def time_to_decimal(time_obj):
    decimal_time = time_obj.hour*3600 + time_obj.minute*60 + time_obj.second*1
    return round(decimal_time, 4)
```

```python
df['start'] = df['time_only_start'].apply(time_to_decimal)
df['end'] = df['time_only_end'].apply(time_to_decimal)
df
```

```python
df.drop('time_only_start', axis=1, inplace=True)
df.drop('time_only_end', axis=1, inplace=True)
df.drop('end', axis=1, inplace=True)
df.drop('Total_Duration_seconds', axis=1, inplace=True)
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score
import skfuzzy as fuzz
from sklearn.mixture import GaussianMixture
import numpy as np
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.metrics import calinski_harabasz_score
```

```python
def compare_clustering_methods(X, k):
    # K-Means
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans_labels = kmeans.fit_predict(X)
    kmeans_cdi = calinski_harabasz_score(X, kmeans_labels)
    kmeans_si = silhouette_score(X, kmeans_labels)
    kmeans_dbi = davies_bouldin_score(X, kmeans_labels)

    # Fuzzy K-Means (Fuzzy C-Means)
    fcm = fuzz.cluster.cmeans(X.T, k, 2, error=0.005, maxiter=1000, init=None)
    fcm_labels = np.argmax(fcm[1], axis=0)
    fcm_cdi = calinski_harabasz_score(X, fcm_labels)
    fcm_si = silhouette_score(X, fcm_labels)
    fcm_dbi = davies_bouldin_score(X, fcm_labels)

    # Gaussian Mixture Model (GMM)
    gmm = GaussianMixture(n_components=k, random_state=42)
    gmm_labels = gmm.fit_predict(X)
    gmm_cdi = calinski_harabasz_score(X, gmm_labels)
    gmm_si = silhouette_score(X, gmm_labels)
    gmm_dbi = davies_bouldin_score(X, gmm_labels)

    return (kmeans_cdi, fcm_cdi, gmm_cdi, kmeans_si, fcm_si, gmm_si, kmeans_dbi, fcm_dbi, gmm_dbi)
```

```python
K = range(2, 11)  # Adjust the range as needed
results = []
X_scaled=df
for k in K:
    metrics = compare_clustering_methods(X_scaled, k)
    results.append((k, *metrics))

results_df = pd.DataFrame(results, columns=['k', 'K-Means CDI', 'Fuzzy K-Means CDI', 'GMM CDI',
                          'K-Means SI', 'Fuzzy K-Means SI', 'GMM SI', 'K-Means DBI', 'Fuzzy K-Means DBI', 'GMM DBI'])
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 18))

# Calinski-Harabasz Index (CDI) plot
ax1.plot(results_df['k'], results_df['K-Means CDI'], 'o-', label='K-Means')
ax1.plot(results_df['k'], results_df['Fuzzy K-Means CDI'],
         'x-', label='Fuzzy K-Means')
ax1.plot(results_df['k'], results_df['GMM CDI'], 's-', label='GMM')
ax1.set_xlabel('Number of clusters (k)')
ax1.set_ylabel('Calinski-Harabasz Index (CDI)')
ax1.set_title('Comparison of Clustering Methods - CDI')
ax1.legend()

# Silhouette Index (SI) plot
ax2.plot(results_df['k'], results_df['K-Means SI'], 'o-', label='K-Means')
ax2.plot(results_df['k'], results_df['Fuzzy K-Means SI'],
         'x-', label='Fuzzy K-Means')
ax2.plot(results_df['k'], results_df['GMM SI'], 's-', label='GMM')
ax2.set_xlabel('Number of clusters (k)')
ax2.set_ylabel('Silhouette Index (SI)')
ax2.set_title('Comparison of Clustering Methods - SI')
ax2.legend()
# Davies-Bouldin Index (DBI) plot
ax3.plot(results_df['k'], results_df['K-Means DBI'], 'o-', label='K-Means')
ax3.plot(results_df['k'], results_df['Fuzzy K-Means DBI'],
         'x-', label='Fuzzy K-Means')
ax3.plot(results_df['k'], results_df['GMM DBI'], 's-', label='GMM')
ax3.set_xlabel('Number of clusters (k)')
ax3.set_ylabel('Davies-Bouldin Index (DBI)')
ax3.set_title('Comparison of Clustering Methods - DBI')
ax3.legend()


plt.tight_layout()
plt.show()
```
